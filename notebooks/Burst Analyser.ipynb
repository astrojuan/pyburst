```python
import numpy as np
import matplotlib.pyplot as plt
from pygrids.burst_analyser import burst_analyser

%matplotlib inline
```

# Loading and analysing models

```python
# Load a kepler model as a BurstRun object
model = burst_analyser.BurstRun(run=1, batch=7, source="biggrid2")

# It will automatically analyse the model for bursts by default.
# To disable this, use the argument 'analyse=False'. 
# You can then run model.analyse() when you wish
```

# Plotting models

```python
# We can plot the whole lightcurve of the model
# The orange points signify detected burst peaks
model.plot_model()
```

```python
# We can also plot the individual burst lightcurves on the same axis.
# Specify bursts by their "index" in the sequence,
# e.g., the first 5 bursts are labelled 0-4
model.plot_lightcurves(bursts=[0, 1, 2, 3, 4])
```

# Model convergence

```python
# An important diagnostic check is whether the burst in a model have "converged"
model.plot_convergence()

# This plots the primary burst properties for each burst in the sequence, 
# and the mean value up to that burst.
# Be careful of your results if there is an obvious trend along the sequence.
# You may need to discard the initial bursts, or run the model for longer
```
